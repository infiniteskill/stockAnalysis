```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
from decimal import Decimal
import numpy as np
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  
```

```python
df=pd.read_csv('Equity.csv')
```

```python
df.sort_values(by=['Security Name'])
```

```python
dfDROP=df['Security Code']
```

```python
dfDROP
```

```python
from bsedata.bse import BSE
b=BSE()
```

```python
datalist=[]
for i in dfDROP:
    try:
        data=b.getQuote(str(i))
        datalist.append([Decimal(i),Decimal(data['dayHigh']),Decimal(data['dayLow']),Decimal(data['52weekHigh']),Decimal(data['52weekLow'])])
        
    except Exception as e:
        print(i,e)
```

```python
b.getQuote(str(500085))
```

```python
MergeR=pd.DataFrame(datalist,columns=['Security Code','dayHigh','dayLow','52weekHigh','52weekLow'])
```

```python
MergeL=df[['Security Code','Security Id','Security Name']]
```

```python
findalDF=pd.merge(MergeL,MergeR,on='Security Code')
```

```python
findalDF=findalDF.sort_values(['dayHigh'])
```
